```python
import xml.etree.ElementTree as ET
from unidecode import unidecode
# parse an xml file by name
file = ET.parse('apertium-fra-eng.fra-eng.dix')

root=file.getroot()
```

```python
def normalize_text(word):
    if word is None:
        return ""
    return unidecode(word).lower()
```

```python
words=set()
for description in root.iter('p'):
    fr_word=description.find('l')
    en_word=description.find('r')
    
    if fr_word.text is not None and normalize_text(fr_word.text)==normalize_text(en_word.text):
        if fr_word.text[0].islower():
            words.add(fr_word.text)
    
print('\n'.join(sorted(words)))
    
```
